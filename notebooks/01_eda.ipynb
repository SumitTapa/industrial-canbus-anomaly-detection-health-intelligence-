# 01 - Industrial CAN Bus EDA

Interactive exploratory data analysis for CAN bus telemetry.

**Note**: The full EDA engine runs as `src/feature_engineering.py`. This notebook is for interactive exploration.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

ROOT = Path('..').resolve()
RAW  = ROOT / 'data' / 'raw'
PROC = ROOT / 'data' / 'processed'
```

```python
df = pd.read_csv(RAW / 'simulated_can_data.csv')
print(f'Records: {len(df):,}')
df.head()
```

```python
fig, axes = plt.subplots(5, 1, figsize=(14, 10), sharex=True)
for ax, col in zip(axes, ['rpm', 'temperature', 'vibration', 'oil_pressure', 'voltage']):
    ax.plot(df[col], lw=0.5)
    ax.set_ylabel(col)
axes[-1].set_xlabel('Time Step')
plt.tight_layout()
plt.show()
```

```python
sensors = ['rpm', 'temperature', 'vibration', 'oil_pressure', 'voltage']
df[sensors].describe()
```
